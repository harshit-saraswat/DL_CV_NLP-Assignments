```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets

%matplotlib inline
```

```python
data=datasets.load_iris()
data.keys()
```

```python
data['target_names']
```

```python
df=pd.DataFrame(data['data'],columns=data['feature_names'])
df['target']=data['target']
df.head()
```

```python
# 2d plot
X = df[['sepal length (cm)','sepal width (cm)']].values
y = df['target'].values

x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5

plt.figure(2, figsize=(8, 6))
plt.clf()
# Plot the training points
plt.scatter(X[:, 0], X[:, 1], c=y,edgecolor='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
```

```python
# 3d plot
fig = plt.figure(figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)
X_reduced = decomposition.PCA(n_components=3).fit_transform(df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)','petal width (cm)']])
ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c=y, edgecolor='k', s=40)
ax.set_title("First 3 PCA Components")
ax.set_xlabel("PCA 1")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("PCA 2")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("PCA 3")
ax.w_zaxis.set_ticklabels([])

plt.show()
```
